# Matplotlib - Subplots Functionality
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2 * x
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
axes
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2)

# for ax in axes:
#     ax.plot(x,y)
axes[0][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][0].plot(x,y)

axes[1][1].plot(a,b)

fig.subplots_adjust(wspace=1,hspace=1)
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(4,10), dpi=200)

# for ax in axes:
#     ax.plot(x,y)
axes[0][0].plot(x,y)

axes[0][1].plot(a,b)

axes[1][0].plot(x,y)
axes[1][0].set_xlabel('X Axis')
axes[1][0].set_ylabel('Y Axis')
axes[1][0].set_xlim(0,6)

axes[1][1].plot(a,b)

fig.suptitle('Figure Level', fontsize=14)

plt.tight_layout()

fig.set_figwidth(10)

fig.savefig('new_subplots.png', bbox_inches='tight')
```
