# Matplotlib - Styling

### 1.1 Styling - Legends
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(-0.25,0.5));
```

### 1.2 Styling - Visual Styling
---

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='#6734eb',label='X v X', lw=5);
ax.plot(x,x**2,color='purple', label='X v X^2',linewidth=2);

ax.legend();
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# ax.plot(x,x,color='#6734eb',label='X v X', lw=5, ls='--');
lines = ax.plot(x,x+1,color='purple', label='X v X+1',linewidth=5);
lines[0].set_dashes([1,2,1,2,5,2])

ax.legend();
```
