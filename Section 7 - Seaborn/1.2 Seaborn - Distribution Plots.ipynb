# Seaborn - Distribution Plots
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(4,3), dpi=200)
sns.rugplot(x='salary', data=df, height=0.5)
```

```python
sns.set(style='white')
sns.displot(data=df, x='salary', bins=20, color='red', 
            edgecolor='black', linewidth=2)
```

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x='salary', color='black',
             edgecolor='yellow', bins=10)
```

```python
plt.figure(figsize=(5,3), dpi=150)
sns.displot(data=df, x='salary', color='orange'
           , edgecolor='black', linewidth=2,
           kde=True, rug=True)
```

```python
sns.kdeplot(data=df, x='salary');
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages, columns=['age'])
```

```python
sample_ages
```

```python
sns.rugplot(data=sample_ages, x='age')
```

```python
sns.displot(data=sample_ages, x='age', rug=True
           ,bins=30, edgecolor='black', color='yellow'
           ,kde=True)
```

```python
sns.kdeplot(data=sample_ages, x='age', clip=[0,100], bw_adjust=0.7, shade=True)
```
