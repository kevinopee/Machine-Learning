# Seaborn - Distributions within Categories
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, y='reading score', 
            x='parental level of education',
           hue='test preparation course')
plt.legend(loc=(1.03, 0.4))
```

```python
plt.figure(figsize=(10,7), dpi=200)
sns.violinplot(data=df, y='parental level of education',
           x='reading score', inner='stick',
              palette='Set2', split=True, bw=0.1)

#plt.legend(loc=(1.03, 0.499))
```

```python
plt.figure(figsize=(8,4), dpi=200)
sns.swarmplot(data=df, x='math score', y='gender', size=2,
             hue='test preparation course')
plt.legend(loc=(1.03, 0.5))
```

```python
sns.boxenplot(data=df, x='math score', 
              y='test preparation course',
              hue='gender')
```
