# Seaborn - Comparison Plots
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('StudentsPerformance.csv')
```

```python
df.head()
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             kind='scatter', color='red', edgecolor='black')
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             kind='kde', color='purple', shade=True,
             hue='gender')
```

```python
sns.jointplot(data=df, x='math score', y='reading score',
             kind='hex', color='black',
             edgecolor='orange')
```

```python
sns.pairplot(data=df, hue='gender', diag_kind='hist',
            palette='Dark2', corner=True)
```
